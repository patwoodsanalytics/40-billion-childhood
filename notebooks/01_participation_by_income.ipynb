```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# CDC / NCHS participation data
# -----------------------------

data = pd.DataFrame({
    "income_group": [
        "Below Poverty Level",
        "100–199% FPL",
        "200–399% FPL",
        "400%+ FPL"
    ],
    "participation_rate": [
        31.2,
        45.0,
        58.0,
        70.2
    ]
})

# -----------------------------
# Create chart
# -----------------------------

plt.figure(figsize=(10, 6))

bars = plt.barh(
    data["income_group"],
    data["participation_rate"]
)

plt.xlabel("Sports Participation Rate (%)")
plt.ylabel("")

plt.title(
    "Youth Sports Participation Rises Dramatically With Family Income",
    pad=20
)

plt.xlim(0, 80)

plt.grid(axis="x", alpha=0.3)

# Add percentage labels
for bar in bars:
    width = bar.get_width()

    plt.text(
        width + 1,
        bar.get_y() + bar.get_height() / 2,
        f"{width:.1f}%",
        va='center'
    )

# Source note
plt.figtext(
    0.01,
    0.01,
    "Source: CDC/NCHS National Health Interview Survey",
    fontsize=9
)

plt.tight_layout()

# Save image
plt.savefig(
    "../visuals/participation_by_income.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
